```python
import numpy as np
import matplotlib.pyplot as plt

#Parametros
lado  = 2

#Creando Rectángulo
vertices = np.array([
    [-lado / 2 - 1j * lado / 2, lado / 2 - 1j * lado / 2],
    [-lado / 2 + 1j * lado / 2, lado / 2 + 1j * lado / 2],
    
])

# Definir la función de traslación compleja f(z) = z + ?
def traslacion(z):
    return z + 2 + 2j

# Aplicar la traslación a los vértices del cuadrado
vertices_trasladados = traslacion(vertices)

# Separar las coordenadas x e y de los vértices trasladados
x_trasladados = np.real(vertices_trasladados)
y_trasladados = np.imag(vertices_trasladados)

#Gráfico del cuadrado original
plt.plot(np.real(vertices), np.imag(vertices), 'b-')
plt.plot(np.append(np.real(vertices), np.real(vertices)[0]), np.append(np.imag(vertices), np.imag(vertices)[0]), 'b-')

# Crear el gráfico del cuadrado trasladado
plt.plot(x_trasladados, y_trasladados, 'r--')
plt.plot(np.append(x_trasladados, x_trasladados[0]), np.append(y_trasladados, y_trasladados[0]), 'r-')

plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.title('Cuadrado Original (Azul) y Cuadrado Trasladado (Rojo) en el Plano Complejo')
plt.axis('equal')
plt.grid(True)
plt.show()
```
